```python
from faker import Faker
```

```python
import pandas as pd 
import random
```

```python
rows=5
data={
 'day':['Mon','Tue','Wed','Thu','Fri'],
    'Pune':[random.randint(20,40) for _ in range(rows)],
   'Satara':[random.randint(20,40) for _ in range(rows)]
    
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df2=df.melt(id_vars='day',var_name='city',value_name='Temp')
df2
```

```python
df3=df2.pivot_table(index='day',columns='city',aggfunc='sum')
df3
```
