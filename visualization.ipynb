```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
heart_data = pd.read_csv('heart.csv')

# 1. Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='chol', data=heart_data)
plt.title('Scatter Plot - Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()

# 2. Histogram
plt.figure(figsize=(8, 6))
sns.histplot(heart_data['age'], bins=10, kde=True)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='thalach', data=heart_data)
plt.title('Box Plot - Sex vs Maximum Heart Rate')
plt.xlabel('Sex')
plt.ylabel('Maximum Heart Rate')
plt.show()

# 4. Bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='cp', y='oldpeak', data=heart_data, estimator=sum)
plt.title('Bar Plot - Chest Pain Type vs Oldpeak')
plt.xlabel('Chest Pain Type')
plt.ylabel('Sum of Oldpeak')
plt.show()

# 5. Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='target', y='trestbps', data=heart_data)
plt.title('Violin Plot - Target vs Resting Blood Pressure')
plt.xlabel('Target')
plt.ylabel('Resting Blood Pressure')
plt.show()

# 6. Pair plot
sns.pairplot(heart_data[['age', 'chol', 'thalach', 'oldpeak']])
plt.title('Pair Plot of Age, Cholesterol, Maximum Heart Rate, and Oldpeak')
plt.show()

# 7. Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heart_data.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()

# 8. Line plot
plt.figure(figsize=(8, 6))
sns.lineplot(x='age', y='thalach', data=heart_data)
plt.title('Line Plot - Age vs Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.show()

# 9. Count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=heart_data)
plt.title('Count Plot - Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# 10. Joint plot
sns.jointplot(x='chol', y='trestbps', data=heart_data, kind='kde')
plt.title('Joint Plot - Cholesterol vs Resting Blood Pressure (Kernel Density Estimation)')
plt.show()

# 11. Swarm plot
plt.figure(figsize=(8, 6))
sns.swarmplot(x='target', y='age', data=heart_data)
plt.title('Swarm Plot - Target vs Age')
plt.xlabel('Target')
plt.ylabel('Age')
plt.show()

# 12. Facet grid
g = sns.FacetGrid(heart_data, col='sex', hue='target', height=5)
g.map(plt.scatter, 'chol', 'trestbps')
g.add_legend()
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Facet Grid - Cholesterol vs Resting Blood Pressure')
plt.show()
```

Differences between Seaborn and Matplotlib:

Ease of Use:
Seaborn: Provides high-level functions for creating informative statistical graphics.
Matplotlib: Offers lower-level control over individual elements of the plot, allowing for more customization but often requiring more code.
Default Styles:
Seaborn: Comes with better default styles and color palettes, making plots aesthetically pleasing.
Matplotlib: Requires more effort to create visually appealing plots due to its basic default styles.
Abstraction Level:
Seaborn: Abstracts away much of the complexity of Matplotlib, making it easier to create complex visualizations with fewer lines of code.
Matplotlib: Provides greater flexibility and control over the appearance of the plots but requires more code for customization.
Statistical Visualization:
Seaborn: Specializes in statistical visualization and offers functions to visualize relationships between variables with ease.
Matplotlib: Provides basic plotting functionality and can be used for a wide range of plotting tasks, from simple bar charts to complex 3D plots.
Integration with Pandas:
Seaborn: Works well with Pandas DataFrame objects and can directly accept DataFrame columns as inputs.
Matplotlib: Can also work with Pandas DataFrame objects, but may require more manual data manipulation compared to Seaborn.


Both libraries have their strengths and are often used together. Seaborn is preferred for statistical visualizations and quick exploration of data, while Matplotlib offers more control and customization options for creating publication-quality plots.
